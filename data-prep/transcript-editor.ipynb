# Transcript Cleaner

```python
import pandas as pd
import numpy as np

import seaborn as sns
sns.set_style("whitegrid")

import matplotlib.pyplot as plt
%matplotlib inline

from google.colab import files

from google.colab import drive
drive.mount('/content/drive')
```

```python
# Insert transcript

path ="/content/drive/MyDrive/Colab Notebooks/Projects/textSearchReplace/captions_text.vtt"
```

```python
vtt = open(path, "r")
txt=vtt.read()
```

```python
# Check content type
# If dict --> separate keys and values

print(type(txt))
```

```python
txt
```

```python
import re
```

```python
# Find timestamps
# Example pattern to look for: \n\n0\n00:00.760 --> 00:05.080\n

pattern = r"\n\n\d{1,3}\n\d\d:\d\d.\d\d\d --> \d\d:\d\d.\d\d\d\n"

regex_time = re.compile(pattern)
regex_time.findall(txt)
```

```python
# Remove timestamps 

unstamped = regex_time.sub(" ", txt)

# Remove  "WEBVTT"

unstamped = unstamped[7:-1]
unstamped
```

```python
len(unstamped)
```

```python
words = unstamped.split()
```

```python
print(type(words))
print(len(words))
words
```

# Word changer

```python
for w in range(len(words)):
  if len(words[w]) == 4 and not words[w][-1] == "." and not words[w][1] == "'" and not words[w][2] == "'":
    words[w] = words[w][0] + "***"
    print(words[w])
```

```python
clean = " ".join(words)
clean
```

```python
with open("clean_transcript.txt", "w") as f:
  f.write(clean)

files.download("clean_transcript.txt")
```

# Scrambler

```python
sample_text = "I have often wondered whether it's possible to read an entire book faster if the useless letters were removed."
```

```python
sample_words = sample_text.split()

sample_words
```

```python
# 100% scrambled

for w in range(len(sample_words)):
  word = sample_words[w]
  idx = np.random.permutation(len(word))
  sample_words[w] = "".join([word[i] for i in idx])

" ".join(sample_words)
```

```python
# Only the mid letters are scrambled

sample_words = sample_text.split()

for w in range(len(sample_words)):
  if len(sample_words[w]) < 4 or "'" in sample_words[w]:
    continue
  
  else:
    word = sample_words[w][1:-1]
    idx = np.random.permutation(len(word))
    sample_words[w] = sample_words[w][0] + "".join([word[i] for i in idx]) \
    + sample_words[w][-1]

" ".join(sample_words)
```

```python
import string
alphabet = string.ascii_lowercase

alphabet
```

```python
# Maximum chaos

sample_words = sample_text.split()

for w in range(len(sample_words)):
  if len(sample_words[w]) < 4 or "'" in sample_words[w]:
    continue
  
  else:
    word = sample_words[w][1:-1]
    idx = np.random.randint(0, len(alphabet), len(word))
    sample_words[w] = sample_words[w][0] + "".join([alphabet[i] for i in idx]) \
    + sample_words[w][-1]

" ".join(sample_words)
```
